### sentiment 라벨링 US

```python
preds = data['preprocessed_pred']
Emoji = {'smile' : 0, 'angry' : 1, 'sob' : 2}

for i, p in enumerate(tqdm(preds)):
    data.loc[i, 'smile_prob'], data.loc[i, 'angry_prob'], data.loc[i, 'sob_prob'] = predictor.predict(p, return_proba = True)
    data.loc[i, 'sentiment'] = predictor.predict(p)
    data.loc[i, 'sentiment_ind'] = Emoji[data.loc[i, 'sentiment']]
```
